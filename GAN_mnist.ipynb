```python
import torch
import torch.nn as nn
import torch.optim as optim
import torchvision.datasets as dsets
import torchvision.transforms as transforms
import torchvision.utils as vutils
from torch.utils.data import DataLoader, Subset
import matplotlib.pyplot as plt
import numpy as np
import os

# ハイパーパラメータの設定
batch_size = 64
image_size = 28    # MNISTは28x28のグレースケール画像
nz = 64           # ノイズベクトルの次元
num_epochs = 100
lr = 0.0002

mnist_label = 8 # 生成したい数字(計算コストの都合上特定の数字に絞って学習)

# 出力画像を保存するディレクトリの作成
os.makedirs("output_images", exist_ok=True)

# MNISTデータセットの前処理：Tensor化と[-1, 1]への正規化
transform = transforms.Compose([
    transforms.ToTensor(),
    transforms.Normalize((0.5,), (0.5,))  # グレースケールなので1チャンネル
])

# MNISTデータセットの読み込み（train=Trueで訓練用データ）
full_dataset = dsets.MNIST(root='./data', train=True, transform=transform, download=True)

# 数字の8だけを抽出する
indices = [i for i, (_, label) in enumerate(full_dataset) if (label == mnist_label)] 
dataset = Subset(full_dataset, indices)
dataloader = DataLoader(dataset, batch_size=batch_size, shuffle=True)

# 生成器（Generator）の定義
class Generator(nn.Module):
    def __init__(self, nz):
        super(Generator, self).__init__()
        self.main = nn.Sequential(
            nn.Linear(nz, 128),
            nn.ReLU(True),
            nn.Linear(128, 256),
            nn.ReLU(True),
            nn.Linear(256, 512),
            nn.ReLU(True),
            nn.Linear(512, image_size * image_size),
            nn.Tanh()  # 出力を[-1, 1]にするため
        )
    
    def forward(self, input):
        output = self.main(input)
        return output.view(-1, 1, image_size, image_size)

# 識別器（Discriminator）の定義
class Discriminator(nn.Module):
    def __init__(self):
        super(Discriminator, self).__init__()
        self.main = nn.Sequential(
            nn.Linear(image_size * image_size, 512),
            nn.LeakyReLU(0.2, inplace=True),
            nn.Linear(512, 256),
            nn.LeakyReLU(0.2, inplace=True),
            nn.Linear(256, 128),
            nn.LeakyReLU(0.2, inplace=True),
            nn.Linear(128, 1),
            nn.Sigmoid()  # 出力を0〜1の確率に
        )
    
    def forward(self, input):
        input_flat = input.view(-1, image_size * image_size)
        output = self.main(input_flat)
        return output

# モデルのインスタンス生成
netG = Generator(nz)
netD = Discriminator()

# 損失関数とオプティマイザの設定
criterion = nn.BCELoss()
optimizerD = optim.Adam(netD.parameters(), lr=lr, betas=(0.5, 0.999))
optimizerG = optim.Adam(netG.parameters(), lr=lr, betas=(0.5, 0.999))

# ラベルの設定
real_label = 1.0
fake_label = 0.0

# 固定ノイズの用意（エポック毎に同じノイズから生成画像を確認するため）
fixed_noise = torch.randn(64, nz)

# GANの学習ループ
for epoch in range(num_epochs):
    for i, (data, _) in enumerate(dataloader):
        batch_size_curr = data.size(0)
        
        ############################
        #  識別器（Discriminator）の更新
        ############################
        netD.zero_grad()
        
        # ① 実際の8の画像データを使って学習
        real_data = data
        label = torch.full((batch_size_curr,), real_label)
        output = netD(real_data).view(-1)
        lossD_real = criterion(output, label)
        lossD_real.backward()
        
        # ② 生成器で生成した偽画像を使って学習
        noise = torch.randn(batch_size_curr, nz)
        fake_data = netG(noise)
        label.fill_(fake_label)
        output = netD(fake_data.detach()).view(-1)
        lossD_fake = criterion(output, label)
        lossD_fake.backward()
        
        lossD = lossD_real + lossD_fake
        optimizerD.step()
        
        ############################
        #  生成器（Generator）の更新
        ############################
        netG.zero_grad()
        # 生成器は識別器を騙して偽画像を本物（8）の画像と判断させたいので、ラベルはrealにする
        label.fill_(real_label)
        output = netD(fake_data).view(-1)
        lossG = criterion(output, label)
        lossG.backward()
        optimizerG.step()
        
        if i % 100 == 0:
            print(f"[Epoch {epoch+1}/{num_epochs}][Batch {i}/{len(dataloader)}] Loss_D: {lossD.item():.4f} Loss_G: {lossG.item():.4f}")
    
    # 10エポックに1回だけ生成画像を出力（表示＆保存）
    if (epoch + 1) % 10 == 0:
        with torch.no_grad():
            fake_images = netG(fixed_noise).detach().cpu()
        grid = vutils.make_grid(fake_images, padding=2, normalize=True)
    
        plt.figure(figsize=(6,6))
        plt.axis("off")
        plt.title(f"Epoch {epoch+1}")
        plt.imshow(np.transpose(grid, (1,2,0)))
        plt.show()
    
        vutils.save_image(fake_images, f"output_images/epoch_{epoch+1:03d}.png", normalize=True)

print("学習終了")
```
